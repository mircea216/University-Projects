```python
#1a)
x=var('x')
y=function('y')(x)
eq1=diff(y,x)-2*y+x^2==0
sol1=desolve(eq1,y)
sol1
```

```python
#1b)
cauchy1=desolve(eq1,y,ics=[0,1/4])
cauchy1
```

```python
#1c)
plot(cauchy1,x,-10,10)
```

```python
#2a) solutie
t=var('t')
x=function('x')(t)
eq2=diff(x,t,2)-6*diff(x,t)+9*x==9*t-6
sol2a=desolve(eq2,x)
sol2a
```

```python
#2a) reprezentare grafica
K1,_K2=var('_K1,_K2')
to_plot=sol2a.substitute(_K1==1,_K2==-1)
to_plot1=sol2a.substitute(_K1==4,_K2==2)
to_plot2=sol2a.substitute(_K1==1,_K2==1)
to_plot3=sol2a.substitute(_K1==5,_K2==-1)
plot([to_plot,to_plot1,to_plot2,to_plot3],t,-1,1)
```

```python
#2b)
a=var('a')
sol2b=desolve(eq2,x,ics=[0,0,a])
sol2b
```

```python
sol2bb=sol2b.substitute(a==1)
sol2bb
# pentru a = 1 solutia e polinomiala
```

```python
#3a)
t=var('t')
x=function('x')(t)
y=function('y')(t)
deq1=diff(x,t)==x+2*y
deq2=diff(y,t)==2*x+y
syst=[deq1,deq2]
_K1,_K2=var('_K1,_K2')
vars=[x,y]
in_cond=[0,_K1,_K2]
solS=desolve_system(syst,vars,in_cond)
solS
```

```python
#3b)
in_cond1=[0,2,0]
cauchy3=desolve_system(syst,vars,in_cond1)
cauchy3
```

```python
#4a)
t,r=var('t,r')
x=function('x')(t)
eq4=diff(x,t)==x*r
sol4=desolve(eq4,[x,t])
sol4
```

```python
#4b) solutie
x0=var('x0')
cauchy4=desolve(eq4,[x,t],ics=[0,x0])
plot_cauchy=cauchy4.substitute(r==2,x0==5)
plot_cauchy
```

```python
#4b) grafic
plot(plot_cauchy,-2,3)
```
