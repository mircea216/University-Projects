```python
x=var('x')
y=function('y')(x)
eqd=diff(y,x,2)+3*diff(y,x)+2*y==1+x^2
desolve(eqd,y)
```

```python
x=var('x')
y=function('y')(x)
eqd1=diff(y,x,2)+diff(y,x)==sin(x)+cos(x)
sol=desolve(eqd1,y)
sol
```

```python
K1,_K2=var('_K1,_K2')
sol.substitute(_K1==1,_K2==3)
sol1=sol.substitute(_K1==1,_K2==3)
plot(sol1,x,-2,2)
```

```python
x=var('x')
y=function('y')(x)
eqd2=diff(y,x,2)-diff(y,x)==e^(2*x)
sol2=desolve(eqd2,y)
sol2
```

```python
K1,_K2=var('_K1,_K2')
sol2.substitute(_K1==1,_K2==3)
sol3=sol2.substitute(_K1==1,_K2==3)
plot(sol3,x,5,10)
```
